## Поставлена задача Коши для решения системы обыкновенных дифференциальных уравнений.

 *Условие задачи*

$$
  \frac{d}{dt}\,v(t) = a*v(t) - \frac{u(t)}{u(t) + 1}\, * b * v(t)
$$


$$
  \frac{d}{dt}\,v(t) = \frac{u(t)}{u(t) + 1}\,v(t) - (u(t) + 1) 
$$

$$t \epsilon\ 0..20 $$
$$a:=4 $$     
$$v(0)=20 $$
$$b:=7$$
$$u(0)=5$$

*Шаги решения*
1. создана функция F(s, t)
    в теле функции записываем переменную dvdt как первое уравнение в векторе s
    переменную dudt записываем в виде второго уравнения которое является вторым элементом вектора S
    возращает функция список состоящий из этих двух переменных
    
2. для нахождения решения в диапозоне решения от 0 до 20 вызывается функция linspace

3. далее задается начальное значение (список). Первый элемент относится к функции v при t = 0, второй элемент относится к функции u при t = 0
                                               
4. Вызываем функцию odeint и подставляем в векторной форме туда функцию, начальное значение и массив значений t а результат вычислений будет определяться ссылкой s, которая ссылается на двумерный массив который состоит из t строк и двух столбцов 0-столбец значения функции v, а первый столбец значение функции u

5. Графическое представление результатов расчета


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def F(s, t):
    a = 4; b = 7
    dvdt = a * s[0] - (s[1]/(s[1]+1)) * b * s[0]
    dudt = (s[1]/(s[1]+1)) * s[0] - s[1] + 1
    return [dvdt, dudt]
t = np.linspace(0, 20)
s0 = [20, 5]
s = odeint(F, s0, t)

plt.plot(t, s[:,0], 'r--', linewidth = 2.0, label = "v(t)")
plt.plot(t, s[:,1], 'b--', linewidth = 2.0, label = "u(t)")
plt.xlabel("t")
plt.ylabel("v(t),    u(t)")
plt.legend()
plt.grid()
plt.show()
```
